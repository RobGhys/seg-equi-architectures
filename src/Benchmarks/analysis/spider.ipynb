```python
labels = ["Relative Performance (Dice)", "Parameter Count", "Data setting"]
num_vars = len(labels)

# Set up the figure
fig, ax = plt.subplots(figsize=(10, 10), subplot_kw=dict(polar=True))

# Calculate the angle for each axis
angles = np.linspace(0, 2 * np.pi, num_vars, endpoint=False).tolist()

# Loop through each model in data
for model, metrics in data.items():
    values = [
        metrics['relative_dice'],  # Example relative performance
        10 if metrics["parameter_count"] == "large" else 5, 
        10 if metrics["datasetting"] == "large" else 5
    ]
    values += values[:1]  # Close the loop for the spider chart

    # Ensure angles also form a closed loop
    angles_full = angles + angles[:1]

    # Plot each model
    ax.plot(angles_full, values, linewidth=1, linestyle='solid', label=model)
    ax.fill(angles_full, values, alpha=0.25)
    
# Customize the axis
ax.set_yticklabels([])
ax.set_xticks(angles)
ax.set_xticklabels(labels)

# Add legend and title with multi-line block format
plt.legend(loc="best", bbox_to_anchor=(1.1, 1.1), ncol=6)  

# Show the plot
plt.show()
```
